### Cleaning Data

Cleaning bike_day

```python
# Hapus variable instant, temp,	atemp,	hum,	windspeed
bike_day.drop("instant", axis=1, inplace=True)
bike_day.drop("temp", axis=1, inplace=True)
bike_day.drop("atemp", axis=1, inplace=True)
bike_day.drop("hum", axis=1, inplace=True)
bike_day.drop("windspeed", axis=1, inplace=True)
```

```python
#Melakukan label encoding untuk mengganti value pada beberapa variabel

season = {1:"Spring", 2:"Summer", 3:"Fall", 4:"Winter"}
yr = {0:2011, 1:2012}
holiday = {0:"not holiday", 1:"holiday"}
weekday = {0:"sunday", 1:"monday", 2:"tuesday", 3:"wednesday", 4:"thursday", 5:"friday", 6:"saturday"}
workingday = {0:"weekend", 1:"not weekend"}
weathersit = {1:"Clear, Few clouds, Partly cloudy, Partly cloudy", 2: "Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist", 3:"Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds", 4:"Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog"}
mnth = {1:"January", 2:"February", 3:"March", 4:"April", 5:"May", 6:"June", 7:"July", 8:"August", 9:"September", 10:"October", 11:"November", 12:"December"}

bike_day["season"]=bike_day["season"].replace(season)
bike_day["yr"]=bike_day["yr"].replace(yr)
bike_day["holiday"]=bike_day["holiday"].replace(holiday)
bike_day["weekday"]=bike_day["weekday"].replace(weekday)
bike_day["workingday"]=bike_day["workingday"].replace(workingday)
bike_day["weathersit"]=bike_day["weathersit"].replace(weathersit)
bike_day["mnth"]=bike_day["mnth"].replace(mnth)
```

```python
# mengubah tipe data menjadi dateday pada variabel dteday
datetime_columns = ["dteday"]
for column in datetime_columns:
    bike_day[column] = pd.to_datetime(bike_day[column])
```

**`Eda bike_day`**

1. Mengetahui Jumlah seluruh penyewa berdasarkan casual, registered, dan total keseluruhan

```python
# Jumlah keseluruhan penyewaan sepeda pada 2011-2012
casual_rental = bike_day["casual"].sum()
registered_rental = bike_day["registered"].sum()
total_rental = bike_day["cnt"].sum()

print("total penyewa tidak terdaftar : ",casual_rental)
print("total penyewa terdaftar : ",registered_rental)
print("total keseluruhan penyewa : ",total_rental)
```

```python
tahun_2011 = bike_day[bike_day['yr'] == 2011]
tahun_2012 = bike_day[bike_day['yr'] == 2012]

total_penyewa_2011 = tahun_2011['cnt'].sum()
total_penyewa_2012 = tahun_2012['cnt'].sum()
print("total penyewa 2011 : ", total_penyewa_2011)
print("total penyewa 2012 : ", total_penyewa_2012)
```

3. Mencari peningkatan penyewaan berdasarkan bulan

```python
# Melihat Jumlah pengunjung berdasarkan bulan
hasil_bulan = bike_day.groupby("mnth").agg({
    "cnt" : ["sum"]})
hasil_bulan
```

4. Mencari peningkatan penyewaan berdasarkan season (musim)


---
Infromasi mengenai musim di USA:
- Winter : Desember – Februari
- Spring : Maret – Mei
- Summer : Juni – Agustus
- Fall : September – November

```python
hasil_musim = bike_day.groupby("season")["cnt"].sum().reset_index()
hasil_musim
```

5. Mencari Jumlah penyewa berdasarkan hari kerja

```python
total_workingday = bike_day.groupby("workingday").agg({"cnt": ["sum"]})
total_workingday
```

6. Mencari Jumlah penyewa berdasarkan hari libur

```python
# Melihat Jumlah pengunjung berdasarkan hari libur
total_holiday = bike_day.groupby("holiday").agg({"cnt": ["sum"]})

total_holiday
```

7. Menggabungkan kolom season dan holiday untuk melihat pola penyewaan

```python
# Menggabungkan kolom season dan holiday, apakah pada liburan di musim tertentu mempengaruhi pola penyewaan
bike_day["season_holiday"] = bike_day["season"].astype(str)+ "_" + bike_day["holiday"].astype(str)
bike_day
```

```python
# Melihat Jumlah pengunjung berdasarkan season_holiday
season_holiday = bike_day.groupby("season_holiday").agg({
    "cnt" : ["sum"]
}).sort_values(("cnt", "sum"), ascending=False)
season_holiday
```
